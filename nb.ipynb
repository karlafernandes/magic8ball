```python
df = pd.read_csv(r"data.csv")
df.drop(columns=['Submission ID', 'Respondent ID', "Submitted at"], inplace=True)
df
```

```python
df["Gender"].unique()
```

```python
df_gender = df["Gender"]
n_male = sum([gender == "Male" for gender in df_gender])
n_female = sum([gender == "Female" for gender in df_gender])
n_other = sum([gender == "Other" for gender in df_gender])
n_male, n_female, n_other
```

```python
import matplotlib.pyplot as plt

labels = ["Female","Male",  "Other"]
sizes = [n_female, n_male, n_other]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, startangle=90, textprops={'fontsize': 12})
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
